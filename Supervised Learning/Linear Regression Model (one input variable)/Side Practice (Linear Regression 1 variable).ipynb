```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import linear_model
```

```python
data=pd.read_csv("F:\canada_per_capita_income.csv")
```

```python
data.tail()
```

```python
# Assuming you have a DataFrame named "data" with the original column name
data.rename(columns={'per capita income (US$)': 'per_capita_income'}, inplace=True)

# Now, the column name is changed to "per_capita_income"
```

```python
years = data['year']
income = data['per_capita_income']
plt.figure(figsize=(10, 6))  # Adjust the figsize as needed
# Create a scatter plot with crosses
plt.scatter(years, income, marker='x', s=50)  # 'x' is the marker style, s is the marker size

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Scatter Plot of Per Capita Income Over Years')

# Show the plot
plt.show()
```

```python
LM= linear_model.LinearRegression()
LM.fit(data[["year"]],data.per_capita_income)
```

```python
import pandas as pd

# Create a dictionary with the 'year' column
da = {'year': [2017, 2018, 2019, 2020]}

# Create the DataFrame
New_data = pd.DataFrame(da)

New_data
```

```python
ba=LM.predict(New_data)
ba
```

```python
New_data["predicted_income"]=ba
```
